```python
dic={'O':10,'S':9,'A':8,'B':7,'C':6,'D':5,'E':4,'F':3,'AB':0}
df['ds']=df['ds'].replace(dic)
df['ds']
```

```python
# Define the mapping dictionary
value_to_number = {'O': 10, 'S': 9, 'A': 8, 'B': 7,'C':6,'D':5,'E':4,'F':3,'AB':0}
df['python'] = df['python'].replace(value_to_number)
df['co'] = df['co'].replace(value_to_number)
print("\nDataFrame with Values Replaced by Numbers:")
print(df)
```

```python
df['os']=df['os'].replace(dic)
```

```python
df['toc']=df['toc'].replace(dic)
```

```python
df['pplab']=df['pplab'].replace(value_to_number)
```

```python
df['oslab']=df['oslab'].replace(value_to_number)
```

```python
df['fosslab']=df['fosslab'].replace(value_to_number)
```

```python
df['matlab']=df['matlab'].replace(value_to_number)
```

```python
df.head()
```

```python
df.isnull().any()
```

```python
# Rename 'OldName' to 'NewName'
df.rename(columns={'python': 'pp'}, inplace=True)
print("\nDataFrame with Renamed Column:")
print(df)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
x=df[['ds','pp','co','toc','os','pplab','oslab','fosslab','matlab']]
y=df['sgpa']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
lm.fit(x,y)
```

```python
ypred=lm.predict(x_test)
print(ypred)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, ypred)
mse = mean_squared_error(y_test, ypred)
r2 = r2_score(y_test, ypred)
print("r2: ",r2)
print("mse:",mse)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lm, x_train, y_train, cv=5)
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

```python
print("mae: ",mae)
```

```python
plt.scatter(y_test, ypred, alpha=0.5)
plt.xlabel('Actual SGPA')
plt.ylabel('Predicted SGPA')
plt.title('Actual vs Predicted SGPA')
plt.show()
```

```python
residuals = y_test - ypred
plt.scatter(ypred, residuals, alpha=0.5)
plt.xlabel('Predicted SGPA')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted SGPA')
plt.axhline(0, color='red', linestyle='--')
plt.show()
```

```python
from sklearn.linear_model import Lasso
la= Lasso(alpha=0.1)  # Alpha is the regularization parameter
la.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Ridge
rd = Ridge(alpha=0.1)  # Alpha is the regularization parameter
rd.fit(x_train, y_train)
```

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet(alpha=0.1, l1_ratio=0.5)  # l1_ratio balances L1 and L2
model.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x, y, cv=5)  # 5-fold cross-validation
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
selector = SelectKBest(score_func=f_regression, k='all')
X_new = selector.fit_transform(x, y)
print("Standard deviation of cross-validation scores:", np.std(scores))
```
