```python
x=df.drop('LUNG_CANCER',axis=1)
y=df['LUNG_CANCER']
```

```python
ypred=dt.predict(x_test)
ypred
```

```python
from sklearn import tree
text_represent=tree.export_text(dt)
print(text_represent)
```

```python
feature_names = x.columns.tolist()  
class_names = ['Yes','No']  
fig = plt.figure(figsize=(25, 20))
tree.plot_tree(dt, feature_names=feature_names, class_names=class_names, filled=True)
fig.savefig('lung_cancer_decision_tree.png')
```

```python
importances = dt.feature_importances_
features = x.columns
feature_importance_df = pd.DataFrame({'Feature': features, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print(feature_importance_df)
```

```python
important_features = feature_importance_df[feature_importance_df['Importance'] > 0.05]['Feature']
X_selected = x[important_features]
```

```python
plt.figure(figsize=(10, 6))
sb.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance')
plt.show()
```

```python
X_train_selected, X_test_selected, y_retrain, y_retest = train_test_split(X_selected, y, test_size=0.25, random_state=42, stratify=y)
```

```python
clf_selected = DecisionTreeClassifier(random_state=42)
clf_selected.fit(X_train_selected, y_retrain)
```

```python
y_pred_selected = clf_selected.predict(X_test_selected)
print(classification_report(y_retest, y_pred_selected))
print(confusion_matrix(y_test, y_pred_selected))
```

```python
plt.figure(figsize=(20, 10))
tree.plot_tree(clf_selected, feature_names=X_selected.columns, class_names=['Yes','No'], filled=True)
plt.savefig('final_decision_tree.png')
plt.show()
```
